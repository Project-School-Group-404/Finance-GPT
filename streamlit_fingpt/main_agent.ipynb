```python
from langchain.tools import tool
from addition import add
from general_qna import gen_qna
from multiplication import multiply

@tool 
def add_tool(a:int , b:int)->int:
    """Add two given numbers"""
    return add(a,b)

@tool 
def gen_qna_tool(question:str)->str:
    """When a general question is asked by the user, this tool should be used to answer that question."""
    return gen_qna(question)


# import sys
# sys.executable
```

```python
from langchain.callbacks.base import BaseCallbackHandler

class PrintEverythingHandler(BaseCallbackHandler):
    def on_llm_new_token(self, token: str, **kwargs) -> None:
        print(token, end="")  # print each token as it arrives

handler = PrintEverythingHandler()
```
